```python
import pandas as pd
import os
import glob
import numpy as np
import re
```

```python
directory_path = "//"
csv_files = glob.glob(os.path.join(directory_path, "*.csv"))

print(f"Found {len(csv_files)} CSV files.")
```

```python
illuminate = pd.read_csv(
    os.path.join(directory_path, "Data/IlluminateData2025.csv"),
    encoding='cp1252',
    low_memory=True
)
```

```python
final_df = pd.read_csv(
    os.path.join(directory_path, "Final_DF.csv"),
    low_memory=True
)
```

```python
illuminate.rename(columns={'Mask_StudentPersonkey': 'student_id'}, inplace=True)
```

```python
for col in illuminate.columns:
    unique_vals = illuminate[col].unique()
    print(f"Column: {col} ({len(unique_vals)} unique values)")
    print(unique_vals[:10])  # Show first 10
    print("-" * 50)
```

```python
for col in final_df.columns:
    unique_vals = final_df[col].unique()
    print(f"Column: {col} ({len(unique_vals)} unique values)")
    print(unique_vals[:10])  # Show first 10
    print("-" * 50)
```

```python
# Ensure both have consistent column name
final_student_ids = final_df['student_id'].astype(str).unique()
illuminate['student_id'] = illuminate['student_id'].astype(str)

# Filter illuminate based on student_id in final_df
illuminate_filtered = illuminate[illuminate['student_id'].isin(final_student_ids)].copy()

# Map 'title' column to standardized 'DepartmentDesc'
title_map = {
    'language arts': 'LANGUAGE ARTS',
    'Language Arts': 'LANGUAGE ARTS',
    'math': 'MATH',
    'social sciences': 'SOCIAL SCIENCES',
    'Science': 'SCIENCE'
}
illuminate_filtered['DepartmentDesc'] = illuminate_filtered['title'].map(title_map)
illuminate_filtered.drop(columns=['title'], inplace=True)
```
